```python
import pandas as pd
import numpy as np
import random
```

```python
df = pd.read_csv("cep-dataset.csv", sep=",", header=0)
df = df.sort_values(by=["PCE"], ascending=False, ignore_index=True)
df
```

```python
sampled_idx = list(range(15)) + [random.randrange(i*15, (i+1)*15) for i in range(1, 1998)]
sampled_df = df.loc[sampled_idx]
sampled_df
```

```python
sampled_df.to_csv("cep-dataset-subsampled.csv", index=False)
```
